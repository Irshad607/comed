```python
rating = [5,4,3,5,2,4,5,3,1,4]
rating_avg = sum(rating)/len(rating)
len(rating)

five_star = 0
for i in rating:
  if i==5:
    five_star+=1

below_avg = 0
for i in rating:
  if i < rating_avg:
    below_avg+=1


print("Average rating is :",rating_avg)
print("Number of users given 5-star",five_star)
print("Days of below average rating",below_avg)
rating.sort(reverse=True)
print(rating)

```

```python
sensor_data = [
    ("Sensor_A", "Field_1", "Temperature", "Celsius"),
    ("Sensor_B", "Field_2", "Humidity", "%"),
    ("Sensor_C", "Field_3", "Soil Moisture", "%"),
    ("Sensor_D", "Field_4", "Temperature", "Celsius"),
    ("Sensor_E", "Field_5", "Humidity", "%")
]
temp_count = 0
for sensor in sensor_data:
  if sensor[2] == "Temperature":
    temp_count+=1
print(f"No of sensors reading temperature is {temp_count}")
for sensor in sensor_data:
  if "Humidity" in sensor:
    humidity_index = sensor.index("Humidity")
    break
print(f"Humidity sensor index is {humidity_index}")
```
